```python
x_train=["jawa is a blockbuster movie",
         "it is a copy pof many movies",
         "music is very worst",
         "nayan looks so beautful"]
y_train=[1,0,0,1] #1-Positive, 0 -Negative
x_train
```

```python
from nltk.tokenize import RegexpTokenizer
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
```

```python
tokenizer=RegexpTokenizer(r'\w+')
en_stopwords=set(stopwords.words('english'))
ps=PorterStemmer()
```

```python
def getCleanedText(text):
  texttext=text.lower()
  tokens=tokenizer.tokenize(text)
  new_tokens=[token for token in tokens if token not in en_stopwords]
  stemmed_tokens=[ps.stem(tokens)for tokens in new_tokens]
  clean_text="".join(stemmed_tokens)
  return clean_text
```

```python
x_test=["hero of that movie is so beautiful", "Rest of the movie were bad"]
x_test
```

```python
x_clean=[getCleanedText(j) for j in x_train]
xt_clean=[getCleanedText(j) for j in x_train]
```

```python
x_clean
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(ngram_range=(1,2))
x_vect=cv.fit_transform(x_clean).toarray()
x_vect
```

```python
print(cv.get_feature_names_out())
```

```python
xt_vect=cv.transform(xt_clean).toarray()
```

```python
from sklearn.naive_bayes import MultinomialNB
mlb=MultinomialNB()
mlb.fit(x_vect,y_train)
y_pred=mlb.predict(xt_vect)
y_pred
```

```python
if y_pred.any()>0 :
  print("positive")
else :
  print("negative")
```
